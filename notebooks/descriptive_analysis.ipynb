```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/acclab_data_innovation.csv')
data[:5]
```

```python
datasources = data.groupby('Datasource')['Lab'].count().sort_values()
y = np.arange(len(datasources.index))
values = list(datasources.values)

plt.barh(y, values, align='center', color = '#0468b1', alpha = 0.5)
plt.yticks(y, list(datasources.index))
plt.xlabel('Count')
plt.title('Datasource')

plt.show()
```
